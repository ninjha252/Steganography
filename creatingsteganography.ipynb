```python
import os
import numpy as np
from PIL import Image

###############################################################################
# 1) INTEGER HAAR 1D (FORWARD & INVERSE)
###############################################################################
def int_haar_1d_forward(signal):
    """
    Integer 1D Haar Transform (lifting-based) on a 1D array of even length.
    Produces approx[] and detail[] as integers.
    """
    n = len(signal)
    half = n // 2
    approx = np.zeros(half, dtype=int)
    detail = np.zeros(half, dtype=int)

    for i in range(half):
        s0 = signal[2*i]
        s1 = signal[2*i + 1]
        d = s1 - s0
        a = s0 + (d // 2)
        approx[i] = a
        detail[i] = d

    return approx, detail

def int_haar_1d_inverse(approx, detail):
    """
    Inverse Integer 1D Haar Transform:
    Reconstructs original signal from approx[] and detail[].
    """
    n = len(approx) + len(detail)
    recon = np.zeros(n, dtype=int)
    for i in range(len(approx)):
        a = approx[i]
        d = detail[i]
        recon[2*i]     = a - (d // 2)
        recon[2*i + 1] = a + d - (d // 2)
    return recon

###############################################################################
# 2) INTEGER HAAR 2D (FORWARD & INVERSE)
###############################################################################
def int_haar_2d_forward(img_array):
    """
    Integer 2D Haar Transform on a grayscale image array.
    Returns sub-bands LL, LH, HL, HH.
    """
    h, w = img_array.shape
    # Row-wise transform
    row_approx  = np.zeros((h, w//2), dtype=int)
    row_detail  = np.zeros((h, w//2), dtype=int)
    for r in range(h):
        a_row, d_row = int_haar_1d_forward(img_array[r, :])
        row_approx[r, :] = a_row
        row_detail[r, :] = d_row

    # Column-wise transform of those halves
    LL = np.zeros((h//2, w//2), dtype=int)
    LH = np.zeros((h//2, w//2), dtype=int)
    HL = np.zeros((h//2, w//2), dtype=int)
    HH = np.zeros((h//2, w//2), dtype=int)
    for c in range(w//2):
        a_col, d_col       = int_haar_1d_forward(row_approx[:, c])
        a_col2, d_col2     = int_haar_1d_forward(row_detail[:, c])
        LL[:, c] = a_col
        HL[:, c] = d_col
        LH[:, c] = a_col2
        HH[:, c] = d_col2

    return LL, LH, HL, HH

def int_haar_2d_inverse(LL, LH, HL, HH):
    """
    Inverse Integer 2D Haar Transform:
    Reconstructs image from sub-bands LL, LH, HL, HH.
    """
    h2, w2 = LL.shape
    h, w   = h2 * 2, w2 * 2

    # Reconstruct the two row-half arrays
    row_approx = np.zeros((h, w2), dtype=int)
    row_detail = np.zeros((h, w2), dtype=int)
    for c in range(w2):
        col0 = int_haar_1d_inverse(LL[:, c], HL[:, c])
        col1 = int_haar_1d_inverse(LH[:, c], HH[:, c])
        row_approx[:, c] = col0
        row_detail[:, c] = col1

    # Reconstruct full image
    out = np.zeros((h, w), dtype=int)
    for r in range(h):
        out[r, :] = int_haar_1d_inverse(
            row_approx[r, :],
            row_detail[r, :]
        )
    return out

###############################################################################
# 3) IWT EMBEDDING & EXTRACTION
###############################################################################
def iwt_embed(cover_path, secret_message, stego_path):
    """
    Embed a UTF-8 text message into the LH sub-band of a grayscale image.
    """
    cover_img = Image.open(cover_path).convert("L")
    w, h = cover_img.size
    new_w, new_h = w + (w % 2), h + (h % 2)
    if (new_w, new_h) != (w, h):
        cover_img = cover_img.resize((new_w, new_h), Image.LANCZOS)
        w, h = new_w, new_h

    arr = np.array(cover_img, dtype=int)
    LL, LH, HL, HH = int_haar_2d_forward(arr)

    # prepare bitstream + end marker
    secret_bytes = secret_message.encode("utf-8")
    secret_bits  = "".join(f"{b:08b}" for b in secret_bytes)
    end_marker   = "1111111111111110"
    bitstream    = secret_bits + end_marker

    flat_LH = LH.flatten()
    if len(bitstream) > flat_LH.size:
        raise ValueError("Message too large for LH capacity.")

    for i, bit in enumerate(bitstream):
        flat_LH[i] = (flat_LH[i] & ~1) | int(bit)

    LH_mod = flat_LH.reshape(LH.shape)
    stego_arr = int_haar_2d_inverse(LL, LH_mod, HL, HH)
    stego_arr = np.clip(stego_arr, 0, 255).astype(np.uint8)

    os.makedirs(os.path.dirname(stego_path) or ".", exist_ok=True)
    Image.fromarray(stego_arr, mode="L").save(stego_path)
    print(f"IWT embedding complete → {stego_path}")

def iwt_extract(stego_path):
    """
    Extract a UTF-8 text message embedded with iwt_embed.
    """
    stego_img = Image.open(stego_path).convert("L")
    w, h = stego_img.size
    new_w, new_h = w + (w % 2), h + (h % 2)
    if (new_w, new_h) != (w, h):
        stego_img = stego_img.resize((new_w, new_h), Image.LANCZOS)

    arr = np.array(stego_img, dtype=int)
    _, LH, _, _ = int_haar_2d_forward(arr)

    bits = "".join(str(val & 1) for val in LH.flatten())
    marker = "1111111111111110"
    idx = bits.find(marker)
    if idx < 0:
        raise ValueError("End marker not found.")
    msg_bits = bits[:idx]
    msg_bytes = [int(msg_bits[i : i+8], 2) for i in range(0, len(msg_bits), 8)]
    return bytes(msg_bytes).decode("utf-8", errors="ignore")

###############################################################################
# 4) LSB EMBEDDING & EXTRACTION
###############################################################################
def lsb_embed(cover_path, secret_message, stego_path):
    """
    Embed a UTF-8 text message into the LSB of an RGB image.
    """
    cover_img = Image.open(cover_path).convert("RGB")
    pixels   = np.array(cover_img)
    flat     = pixels.flatten()

    secret_bytes = secret_message.encode("utf-8")
    bits         = "".join(f"{b:08b}" for b in secret_bytes)
    end_marker   = "1111111111111110"
    bitstream    = bits + end_marker

    if len(bitstream) > flat.size:
        raise ValueError("Message too large for LSB capacity.")

    for i, bit in enumerate(bitstream):
        flat[i] = (flat[i] & ~1) | int(bit)

    new_pixels = flat.reshape(pixels.shape)
    os.makedirs(os.path.dirname(stego_path) or ".", exist_ok=True)
    Image.fromarray(new_pixels).save(stego_path)
    print(f"LSB embedding complete → {stego_path}")

def lsb_extract(stego_path):
    """
    Extract a UTF-8 text message embedded with lsb_embed.
    """
    stego_img = Image.open(stego_path).convert("RGB")
    flat      = np.array(stego_img).flatten()

    bits       = "".join(str(val & 1) for val in flat)
    marker     = "1111111111111110"
    idx        = bits.find(marker)
    if idx < 0:
        raise ValueError("End marker not found.")
    msg_bits   = bits[:idx]
    msg_bytes  = [int(msg_bits[i : i+8], 2) for i in range(0, len(msg_bits), 8)]
    return bytes(msg_bytes).decode("utf-8", errors="ignore")
```

```python
%%time 


import os
import random
import pandas as pd
from pathlib import Path
from PIL import Image

# ──────────────────────────────────────────────────────────────────────────────
# ASSUMPTIONS: 
#  • iwt_embed(), lsb_embed() are defined above
#  • STAR_QUOTES = [ … ] is defined above
# ──────────────────────────────────────────────────────────────────────────────

INPUT_DIR  = "./Images"
OUTPUT_DIR = "stego_images"
CSV_DIR    = "csv"

# gather all image files
all_files = [p for p in Path(INPUT_DIR).iterdir() 
             if p.suffix.lower() in {".jpg", ".jpeg", ".png", ".bmp"}]
total = len(all_files)

# build method list
n_iwt  = int(0.30 * total)
n_lsb  = int(0.40 * total)
n_none = total - n_iwt - n_lsb

methods = ["iwt"] * n_iwt + ["lsb"] * n_lsb + ["none"] * n_none
random.shuffle(methods)

# prepare output folder
os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(CSV_DIR,    exist_ok=True)

# counters for naming
counters = {"iwt": 1, "lsb": 1, "none": 1}
records  = []

for img_path, method in zip(all_files, methods):
    orig = str(img_path)
    ext  = img_path.suffix.lower()
    
    if method == "iwt":
        secret = random.choice(STAR_QUOTES)
        fname  = f"iwt_img{counters['iwt']}{ext}"
        out    = os.path.join(OUTPUT_DIR, fname)
        iwt_embed(orig, secret, out)
        counters["iwt"] += 1

    elif method == "lsb":
        secret = random.choice(STAR_QUOTES)
        fname  = f"lsb_img{counters['lsb']}{ext}"
        out    = os.path.join(OUTPUT_DIR, fname)
        lsb_embed(orig, secret, out)
        counters["lsb"] += 1

    else:  # none
        secret = ""
        fname  = f"none_img{counters['none']}{ext}"
        out    = os.path.join(OUTPUT_DIR, fname)
        Image.open(orig).save(out)
        counters["none"] += 1

    records.append({
        "original_path":   orig,
        "new_image_path":  out,
        "secret_message":  secret,
        "method":          method
    })

# write the CSV
df = pd.DataFrame(records,
                  columns=["original_path","new_image_path","secret_message","method"])
df.to_csv(os.path.join(CSV_DIR, "stego_images.csv"), index=False)

print(f"Done: processed {total} images → {OUTPUT_DIR}/ + {CSV_DIR}/stego_images.csv")
```

```python
import os
import shutil
import pandas as pd
from pathlib import Path

CSV_IN      = "./csv/stego_images.csv"
STEG_DIR    = Path("stego_images")
SUCCESS_DIR = Path("successful")
CSV_OUT     = "./csv/successful_stego_images.csv"

# ensure the success dir exists
SUCCESS_DIR.mkdir(exist_ok=True, parents=True)

# load the original CSV
df = pd.read_csv(CSV_IN)

records = []
moved   = 0
skipped = 0

for _, row in df.iterrows():
    orig   = row["original_path"]
    new    = Path(row["new_image_path"])
    method = row["method"]
    secret = row["secret_message"] or ""
    
    # skip missing files
    if not new.exists():
        print(f"[MISSING] {new} does not exist; skipping.")
        skipped += 1
        continue
    
    should_move = False

    if method == "none":
        # always move the none images
        should_move = True

    elif method == "iwt":
        try:
            extracted = iwt_extract(str(new))
            if extracted == secret:
                should_move = True
            else:
                print(f"[MISMATCH IWT] {new.name}: expected {secret!r}, got {extracted!r}")
        except Exception as e:
            print(f"[ERROR IWT] {new.name}: {e}")

    elif method == "lsb":
        try:
            extracted = lsb_extract(str(new))
            if extracted == secret:
                should_move = True
            else:
                print(f"[MISMATCH LSB] {new.name}: expected {secret!r}, got {extracted!r}")
        except Exception as e:
            print(f"[ERROR LSB] {new.name}: {e}")

    if should_move:
        dst = SUCCESS_DIR / new.name
        shutil.move(str(new), str(dst))
        records.append({
            "original_path":     orig,
            "final_image_path":  str(dst),
            "method":            method,
            "secret_message":    secret
        })
        moved += 1
    else:
        skipped += 1

# write out the new CSV
pd.DataFrame(records).to_csv(CSV_OUT, index=False)

print(f"\nDone. Moved {moved} files to '{SUCCESS_DIR}/', skipped {skipped}.")
print(f"New CSV written to: {CSV_OUT}")
```
