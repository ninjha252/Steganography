```python
import os
import numpy as np
from PIL import Image

def lsb_embed(cover_path, secret_message, stego_path):
    """
    Embed a UTF-8 text message into the LSB of the BLUE channel of an RGB image.
    Uses a 32-bit header for message length. Saves as PNG to avoid lossy compression.
    """
    # Load and get blue channel
    img    = Image.open(cover_path).convert("RGB")
    pixels = np.array(img, dtype=np.uint8)
    h, w, _ = pixels.shape
    blue    = pixels[:, :, 2].flatten()

    # Prepare bitstream: 32-bit length header + message bits
    data       = secret_message.encode("utf-8")
    bit_len    = len(data) * 8
    header     = f"{bit_len:032b}"
    payload    = "".join(f"{byte:08b}" for byte in data)
    bitstream  = header + payload

    # Capacity check: header+payload must fit in h*w pixels
    if len(bitstream) > h * w:
        raise ValueError(f"Message too large ({len(bitstream)} bits) for capacity ({h*w} bits).")

    # Embed into LSB of blue channel
    for i, bit in enumerate(bitstream):
        blue[i] = (blue[i] & 0xFE) | int(bit)

    # Write back and save lossless
    pixels[:, :, 2] = blue.reshape((h, w))
    os.makedirs(os.path.dirname(stego_path) or ".", exist_ok=True)
    Image.fromarray(pixels).save(stego_path, format="PNG")
    print(f"LSB embedding complete → {stego_path}")

def lsb_extract(stego_path):
    """
    Extract a UTF-8 text message embedded with lsb_embed() above.
    Reads the 32-bit length header, then that many bits from the LSB of the blue channel.
    """
    img    = Image.open(stego_path).convert("RGB")
    pixels = np.array(img, dtype=np.uint8)
    h, w, _ = pixels.shape
    blue    = pixels[:, :, 2].flatten()

    # Read 32-bit header
    header_bits = "".join(str(blue[i] & 1) for i in range(32))
    bit_len     = int(header_bits, 2)

    # Extract payload bits
    bits = "".join(str(blue[i] & 1) for i in range(32, 32 + bit_len))

    # Reconstruct bytes
    data = bytearray()
    for i in range(0, len(bits), 8):
        byte = bits[i : i+8]
        data.append(int(byte, 2))

    return data.decode("utf-8", errors="ignore")
```

```python
import os
import random
import shutil
import pandas as pd
from pathlib import Path

# ──────────────────────────────────────────────────────────────────────────────
# ASSUMPTIONS:
#  • STAR_QUOTES list is defined above
#  • lsb_embed() and lsb_extract() are defined/imported above
# ──────────────────────────────────────────────────────────────────────────────

IMG_DIR     = Path("Images")
INITIAL_DIR = Path("initial")
STEGO_DIR   = Path("stego")
CSV_PATH    = Path("succ_stego.csv")

# 1) Prepare folders
INITIAL_DIR.mkdir(exist_ok=True)
STEGO_DIR.mkdir(exist_ok=True)

# 2) Sample 2 000 images into initial/
all_imgs = [p for p in IMG_DIR.iterdir() 
            if p.suffix.lower() in {".jpg",".jpeg",".png",".bmp"}]
sampled = random.sample(all_imgs, min(2000, len(all_imgs)))
for src in sampled:
    shutil.copy(src, INITIAL_DIR / src.name)

# 3) From those, pick 1 000 to LSB‐embed, rename them to lsb_{i}
to_encode = random.sample(sampled, min(1000, len(sampled)))

records = []
counter = 1

for orig in to_encode:
    ext           = orig.suffix.lower()
    initial_copy  = INITIAL_DIR / orig.name
    renamed_copy  = INITIAL_DIR / f"lsb_{counter}{ext}"

    # rename the initial copy so we know which have been used
    initial_copy.rename(renamed_copy)

    # embed into stego/lsb_{i}.ext
    secret     = random.choice(STAR_QUOTES)
    stego_path = STEGO_DIR / f"lsb_{counter}{ext}"
    lsb_embed(str(renamed_copy), secret, str(stego_path))

    # verify extraction
    try:
        extracted = lsb_extract(str(stego_path))
        if extracted == secret:
            records.append({
                "original_path":   str(orig),
                "stego_path":      str(stego_path),
                "method":          "lsb",
                "secret_message":  secret
            })
        else:
            print(f"[MISMATCH] {stego_path.name}: expected {secret!r}, got {extracted!r}")
            stego_path.unlink()
    except Exception as e:
        print(f"[ERROR] decoding {stego_path.name}: {e}")
        if stego_path.exists():
            stego_path.unlink()

    counter += 1

# 4) Write summary CSV
pd.DataFrame(records).to_csv(CSV_PATH, index=False)

print(f"Done! {len(records)} LSB images verified and moved to '{STEGO_DIR}/'.")
print(f"Log written to '{CSV_PATH}'.")
```
